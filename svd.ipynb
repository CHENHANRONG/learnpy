# learning Singular Value Decomposion

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from PIL import Image
plt.style.use('classic')
img = Image.open('images/lucy_goose.jpg')
# convert image to grayscale
imggray = img.convert('LA')
# convert to numpy array
imgmat = np.array(list(imggray.getdata(band=0)), float)
print(imgmat.shape)

# Reshape according to orginal image dimensions
imgmat.shape = (imggray.size[1], imggray.size[0])

plt.figure(figsize=(9, 6))
plt.imshow(imgmat, cmap='gray')
plt.show()
```

```python
U, D, V = np.linalg.svd(imgmat)
print(imgmat.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

```python
for n in (5,10,25,50, 100,200,250):
    print("n="+str(n))
    # SVD reconstruct the img
    reconstruct_img = np.matrix(U[:, :n]) * np.diag(D[:n]) * np.matrix(V[:n, :])
    plt.imshow(reconstruct_img, cmap='gray')
    plt.show()
```
